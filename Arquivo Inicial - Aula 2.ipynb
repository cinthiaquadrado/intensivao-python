```python
import pandas as pd

tb = pd.read_csv('telecom_users.csv')

tb = tb.drop('Unnamed: 0', axis=1) #axis 1 é para a coluna e axis 0 é para a linha

tb
```

```python
tb['TotalGasto'] = pd.to_numeric(tb['TotalGasto'], errors="coerce")

tb = tb.dropna(how="all", axis=1) #excluir colunas em que todos os valores são vazios

tb = tb.dropna(how="any", axis=0) #excluir linhas que tem pelo menos 1 valor vazio (que possuem ALGUM valor vazio)
```

```python
tb['Churn'].value_counts()
tb['Churn'].value_counts(normalize=True).map("{:.1%}".format)
```

```python
tb.info() #informações que estão nos dados
```

```python
import plotly.express as px

#comparar cada coluna da base de dados com a coluna Churn
#cria gráfico
#exibe gráfico

#para cada coluna da minha tabela, cria-se um novo gráfico
for coluna in tb.columns:
    grafico = px.histogram(tb, x=coluna, color="Churn")
    grafico.show()

#link para visualizar os gráficos online
#https://colab.research.google.com/drive/1mlTJQ5GpACcSr-2wo09cmcbg4x7uFrTR?usp=sharing
```
